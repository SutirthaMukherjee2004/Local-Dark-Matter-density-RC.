```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import numpy as np
import pandas as pd
Rsun=8
def calculate_R(result, Rsun=8):
    r = 1 / result['parallax']
    X = Rsun - (r * np.cos(result['l']) * np.cos(result['b']))
    Y = r * np.sin(result['l']) * np.cos(result['b'])
    R = np.sqrt(X*X + Y*Y)
    return R


def calculate_vl(result, U_sun=11.1, V_sun=245.8):
 
    dec_G, ra_G = np.radians(27), np.radians(193)
    C1 = (np.sin(dec_G) * np.cos(result['dec']) - np.cos(dec_G) * np.sin(result['dec']) * np.cos(result['ra'] - ra_G))
    C2 = (np.cos(dec_G) * np.sin(result['ra'] - ra_G))
    mul = (1 / np.cos(result['b'])) * ((C1 * result['pmra']) + (C2 * result['pmdec']))
    vl=mul/result['parallax']
    return vl
    
def calculate_Vc(result, U_sun=11.1, V_sun=245.8):
    r = 1 / result['parallax']
    X = Rsun - (r * np.cos(result['l']) * np.cos(result['b']))
    Y = r * np.sin(result['l']) * np.cos(result['b'])
    phi = np.arctan(Y/X)
    y1 = (calculate_vl(result) - (U_sun * np.sin(result['l'])) + (V_sun * np.cos(result['l']))) / (np.cos(phi + result['l']))
    return y1


x1 = calculate_R(results)
y1 =calculate_Vc(results)

x2 = calculate_R(results1)
y2 =calculate_Vc(results1)

x3= calculate_R(results2)
y3 =calculate_Vc(results2)

# Function to compute binned statistics
def compute_binned_stats(x, y, n_bins=900):
    data = pd.DataFrame({'x': x, 'y': y})
    data['bin'] = pd.cut(data['x'], bins=n_bins, labels=False)
    binned_data = data.groupby('bin').agg({
        'x': 'mean',
        'y': ['mean', 'sem']
    }).reset_index()
    binned_data.columns = ['bin', 'x_mean', 'y_mean', 'y_sem']
    return binned_data

# Compute binned statistics for both datasets
binned_data1 = compute_binned_stats(x1, y1)
binned_data2 = compute_binned_stats(x2, y2)
binned_data3 = compute_binned_stats(x3, y3)

# Plotting
plt.figure(figsize=(12, 8))

# Unbinned scatter plots
plt.subplot(2, 1, 1)
plt.scatter(x3, y3, color='r', alpha=0.5, label='Dataset 2')
plt.scatter(x1, y1, color='b', alpha=0.5, label='Dataset 1')
plt.scatter(x2, y2, color='g', alpha=0.5, label='Dataset 2')
plt.title('Unbinned Scatter Plots')
plt.xlabel('X')
plt.xlim(0, 20)
plt.ylabel('Y')
plt.ylim(-500,500)
plt.legend()

# Binned scatter plots
plt.subplot(2, 1, 2)
plt.errorbar(binned_data1['x_mean'], binned_data1['y_mean'], yerr=binned_data1['y_sem'], fmt='o', color='b', label='Dataset 1')
plt.plot(binned_data1['x_mean'], binned_data1['y_mean'], color='b', linestyle='-', linewidth=1)


plt.errorbar(binned_data2['x_mean'], binned_data2['y_mean'], yerr=binned_data2['y_sem'], fmt='o', color='g', label='Dataset 2')
plt.plot(binned_data2['x_mean'], binned_data2['y_mean'], color='g', linestyle='-', linewidth=1)

plt.errorbar(binned_data3['x_mean'], binned_data3['y_mean'], yerr=binned_data3['y_sem'], fmt='o', color='r', label='Dataset 3')
plt.plot(binned_data3['x_mean'], binned_data3['y_mean'], color='r', linestyle='-', linewidth=1)
plt.title('Binned Scatter Plots')
plt.xlabel('X (Binned)')
plt.xlim(0, 20)
plt.ylabel('Y (Mean)')
plt.ylim(-500,700)
plt.legend()

'''plt.tight_layout()
from scipy.interpolate import UnivariateSpline

def plot_binned_data_with_smooth_line(x_bins_list, y_means_list, y_errors_list, colors=None):
    plt.figure(figsize=(10, 6))
    
    # Iterate over each dataset
    for i, (x_bins, y_means, y_errors) in enumerate(zip(x_bins_list, y_means_list, y_errors_list)):
        if colors:
            color = colors[i]
        else:
            color = None
        
        # Sort data by x_bins to ensure proper spline interpolation
        sort_indices = np.argsort(x_bins)
        x_bins_sorted = x_bins[sort_indices]
        y_means_sorted = y_means[sort_indices]
        y_errors_sorted = y_errors[sort_indices]
        
        # Plot binned data with error bars
        plt.errorbar(x_bins_sorted, y_means_sorted, yerr=y_errors_sorted, fmt='o', color=color, label=f'Dataset {i+1} - Binned Data with Error Bars')
        
        # Fit a smooth regression line (spline interpolation)
        spline = UnivariateSpline(x_bins_sorted, y_means_sorted, w=1/y_errors_sorted,s=10)
        x_smooth = np.linspace(x_bins_sorted.min(), x_bins_sorted.max(), 100)
        y_smooth = spline(x_smooth)
        # Plot smooth regression line
        plt.plot(x_smooth, y_smooth, color=color, linestyle='-', linewidth=2, alpha=0.8, label=f'Dataset {i+1} - Smooth Regression Line')
    
    plt.xlabel('X')
    plt.xlim(0, 20)
    plt.ylabel('Y')
    plt.ylim(-500,900)
    plt.title('Binned Data with Smooth Regression Lines')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_binned_data_with_smooth_line([np.array(binned_data1['x_mean']), 
                                   np.array(binned_data2['x_mean']), np.array(binned_data3['x_mean'])],
                                   
                                    [np.array(binned_data1['y_mean']), np.array(binned_data2['y_mean']), np.array(binned_data3['y_mean'])], 
                                    
                                    [np.array(binned_data1['y_sem']), np.array(binned_data2['y_sem']), np.array(binned_data3['y_sem'])], 
                                    
                                    colors=['b', 'g','r'])

'''
```

```python
from astroquery.sdss import SDSS
from astroquery.sdss import SkyServer
import pandas as pd
from astropy.coordinates import SkyCoord
import astropy.units as u

# Define a function to query the SDSS for red clump giants
def query_red_clump_giants(ra, dec, radius=1.0):
    """
    Query SDSS for red clump giants in a given region of the sky.
    
    Parameters:
    - ra (float): Right Ascension (degrees).
    - dec (float): Declination (degrees).
    - radius (float): Search radius (arcminutes).
    
    Returns:
    - pandas DataFrame with SDSS star data.
    """
    # Define the coordinates (RA, DEC)
    coord = SkyCoord(ra=ra, dec=dec, unit=(u.deg, u.deg), frame='icrs')
    
    # Access SDSS via Astroquery
    xid = SkyServer.query_region(coord, radius=radius*u.arcmin, data_release=12)  # Use data release 12 for SDSS
    
    # Filter for red clump giants using specific color and magnitude ranges
    # This is a simplistic filter based on typical parameters for red clump giants
    red_clump_giants = xid[xid['u'] - xid['g'] < 1.0]  # Example color filter (u-g < 1.0 for red clump stars)
    red_clump_giants = red_clump_giants[red_clump_giants['g'] < 18]  # Example magnitude filter (g < 18)
    
    return red_clump_giants

# Example usage:
ra = 180.0  # Right Ascension in degrees
dec = 0.0   # Declination in degrees

# Query SDSS for red clump giants around a specific RA, Dec position
red_clump_data = query_red_clump_giants(ra, dec)

# Show results in a pandas DataFrame
print(red_clump_data)
```
